Imputation by Mean, Median, LOCF, NOCB, Rolling_Statistics, Interpolation

```python
def Mean_imputation(df):
  df = df[df.Year>1979]
  df.fillna(df.mean(),inplace=True)
  return df
def Median_imputation(df):
  df = df[df.Year>1979]
  df.fillna(df.median(),inplace=True)
  return df
def LOCF(df):
  df = df[df.Year>1979]
  df=df.ffill()
  return df
def NOCB(df):
  df = df[df.Year>1979]
  df=df.bfill()
  return df
def Rolling_Statistics(df):
  df = df[df.Year>1979]
  d_f=df.iloc[:,1:].ewm(halflife=4).mean()
  d_f.insert(0,"Year",df.iloc[:,0])
  return d_f
def interpolation(df):
  df = df[df.Year>1979].astype(float)
  df=df.interpolate('linear',axis=0)
  return df


Missing_technique=[" Mean", " Median", " LOCF", " NOCB"," Rolling statistics"," Interpolate"]
```

```python
def create_df_output(df,Country_code,Missing_technique,k):
  if k ==1:
    dataset='Manufacturing_filled.xlsx'
  elif k==0:
    dataset="Service_filled.xlsx"
  with pd.ExcelWriter(dataset, engine='xlsxwriter') as writer:
    for x in range(len(Country_code)):
      for y in range(len(Missing_technique)):
        if y == 0:
          df_temp = Mean_imputation(df[x])
        elif y == 1:
          df_temp = Median_imputation(df[x])
        elif y == 2:
          df_temp = LOCF(df[x])
        elif y == 3:
          df_temp = NOCB(df[x])
        elif y == 4:
          df_temp = Rolling_Statistics(df[x])
        elif y == 5:
          df_temp = interpolation(df[x])
         
        df_temp.to_excel(writer, sheet_name=Country_code[x]+Missing_technique[y], index=False)
 
```
